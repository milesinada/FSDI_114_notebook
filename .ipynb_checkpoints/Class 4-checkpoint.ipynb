```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        
    def append(self, item):
        if not self.head:
            self.head = Node(item)
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = Node(item)
            
    def insert(self, index, value):
        new_node = Node(value)
        if not self.head:
            self.head = new_node
        else:
            if index == 0:
                temp_node = self.head
                self.head = new_node
                new_node.next = temp_node
            else:
                counter = 0
                prev_node = None
                current = self.head
                while current.next and index != counter:
                    counter += 1
                    prev_node = current
                    current = current.next
                if not current.next:
                    current.next = new_node
                else:
                    prev_node.next = new_node
                    new_node.next = current
    
    def counter(self, value):
        current = self.head
        count = 0
        while current != None:
            if current.data == value:
                count += 1
            current = current.next
        return count
    
    
    def remove(self, value):
        current = self.head
        if current.data == value:   #This is O(1)
            self.head = current.next
        else:
            prev_node = None
            while current.next and current.data != value: #worst-case (Avg-case) is O(n)
                prev_node = current
                current = current.next
            if current.data == value:
                prev_node.next = current.next
            else:
                raise ValueError("%s not is list" % repr(value))
                
                
    def __len__(self):
        counter = 0
        current = self.head
        while current:
            counter += 1
            current = current.next
        return counter
    
    def __getitem__(self, index):
        counter = 0
        current = self.head
        while current and index != counter:
            counter += 1
            current = current.next
        if not current:
            raise IndexError("list index not in range")
        return current.data
    
        
        
    def __str__(self):
        out_str = "["
        current = self.head
        while current.next:
            out_str += "%s, " % repr(current.data)
            current = current.next
        out_str += "%s]" % str(current.data)
        return out_str
    
```

```python
class Dnode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
    def __str__(self):
        if self.data:
            return self.data__str__()
        else:
            return 'Empty Node'
        
class DoublyLinkedList:
    def __init__(self):
        self.head == None
        
    def append(self, value):
        new_node = Dnode(value)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current
            
    def insert(self, prev_node, value):
        if prev_node == None:
            print("No previous node.")
            return
        new_node = Dnode(value)
        new_node.next = prev_node.next
        prev_node.next = new_node
        new_node.prev = prev_node
        if new_node.next:
            new_node.next.prev = new_node
    
    def remove(self, value):
        if self.head == None:
            return
        if self.head == value:
            self.head = value.next
        if value.next != None:
            value.next.prev = value.prev
        if value.prev != None:
            value.prev.next = value.next
    
# Your dll needs : insert, remove, and __str__ methods
```
